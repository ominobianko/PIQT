```
import pandas as pd
import os  
from pathlib import Path
from bs4 import BeautifulSoup
```

```
#fuznione estrai titolo per definizione test 
def titoli(ffile, tag):
    t=[]
    f =  Path(ffile).read_text()
    soup=BeautifulSoup(f)
    a=soup.find_all(tag)
    for x in a:
        b=x.getText()
        t.append(b) 
    #print(t)
    return (t)
```

```
#funzione estrai deal file f le tabelle HTML e seleziona quelle con in dati utili

def estraitabelle (f):

    tabelleHTML = pd.read_html(f,header=None)
    #tabelleHTML.drop(columns=a.columns[-1:], axis=1,  inplace=True)
    selected=[]
    for i in [2, 5, 8, 11]:
        temp=[tabelleHTML[i]]
        selected=selected+temp
    return selected
```

```
def Tabella (sergente, x):
    #C Ã¨ la lista dei dataframe estratti
    C=[]
    a=estraitabelle(sergente)
    b=titoli(sergente,'span')
    A=a[x].transpose()
    A.columns = A.iloc[0]
    A = A[1:]
    A.insert(0,'QATest',b[x])
    temp=[A]
    C=C+temp
    #print(a)
    return C
```

```
#Analizza la cartella 
import os  

cartella= 'user'

m=os.listdir(cartella)
fuori=[]
for x in range (0, 4):

    #print (len(m))
    for n in range(0, len(m)):
        path=cartella+'/'+str(m[n])
        fuori=fuori+Tabella(path,x)
```

```
#concatena tutte le tabelle estratte con funzione tabella in un unico dataframe

out=[]
out = pd.concat(fuori, axis=0, ignore_index=True)
out.drop_duplicates()
out.to_csv('out.csv',index=False,)
```
